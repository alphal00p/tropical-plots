```python
import matplotlib.pyplot as plt
import matplotlib.cm as cms
from scipy import stats
import json
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
plt.rcParams['text.usetex'] = True
plt.rc('font', family='serif', size=14)
```

```python
x_file_triangle = open("triangle_2d_0.json", "r").read()
y_file_triangle = open("triangle_2d_1.json", "r").read()
x_data_triangle = np.array(json.loads(x_file_triangle))
y_data_triangle = np.array(json.loads(y_file_triangle))
```

```python
x_file_double_triangle = open("double_triangle_0.json", "r").read()
y_file_double_triangle = open("double_triangle_1.json", "r").read() 

x_file_data = np.array(json.loads(x_file_double_triangle)) 
y_file_data = np.array(json.loads(y_file_double_triangle)) 
```

```python
fig, ax = plt.subplots(1,1) 
my_cmap = cms.magma.reversed()
hist = ax.hist2d(x_file_data, y_file_data, bins=100, range=[[-3,3],[-3,3]], density=True, norm='log', vmin=1.0e-4, vmax=10.0, cmap=my_cmap, rasterized=True)
#ax.set_title(r'Double triangle')
ax.set(xlabel=r'$k$')
ax.set(ylabel=r'$l$')
fig.colorbar(hist[3], label='Sampling Density')
plt.savefig('double_triangle.pdf', format='pdf', dpi=300) 
```
